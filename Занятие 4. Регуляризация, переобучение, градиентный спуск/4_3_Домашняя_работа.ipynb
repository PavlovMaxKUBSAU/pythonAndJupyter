Обучите модель `sklearn.linear_model.Lasso` и модель `sklearn.linear_model.Ridge` на полученных данных, используйте коэффициент регуляризации $\alpha=0.8$ для обеих моделей. Постройте два столбчатых графика, на которых отобразите величину коэффициентов для Ridge регрессии и Lasso регрессии в виде столбиков.

Пример графиков для 3-й степени (просто для примера, у вас может по-другому выглядеть). Какой можно сделать в вывод по величине коэффициентов?:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
#x = np.arange(-5.0, 5.0, 0.1)
#plt.figure(figsize=(10,5))
#y = 1*(x**3) + 1*(x**2) + 1*x + 3
#y_noise = 20 * np.random.normal(size=x.size)
#ydata = y + y_noise
#plt.plot(x,y, 'r') 
#plt.plot(x, ydata,  'bo')
#plt.ylabel('y')
#plt.xlabel('x')
#plt.show()

def generate_degrees(source_data: list, degree: int):
    """Функция, которая принимает на вход одномерный массив, а возвращает n-мерный
    Для каждой степени от 1 до  degree возводим x в эту степень
    """
    return np.array([
          source_data**n for n in range(1, degree + 1)
    ]).T




data = pd.read_csv('non_linear.csv', sep=',')
degree = 8
X = generate_degrees(data['x_train'], degree)
y = data.y_train.values
#X = generate_degrees(x, degree)
#y = ydata
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=10)

#Ridge 0
model1 = Ridge(alpha=0).fit(X_train, y_train)
y_pred_train1 = model1.predict(X_train)
y_pred_valid1 = model1.predict(X_valid)
error_train1 = mean_squared_error(y_train, y_pred_train1)
error_valid1 = mean_squared_error(y_valid, y_pred_valid1)

#Ridge 0.8
model2 = Ridge(alpha=0.8).fit(X_train, y_train)
y_pred_train2 = model2.predict(X_train)
y_pred_valid2 = model2.predict(X_valid)
error_train2 = mean_squared_error(y_train, y_pred_train2)
error_valid2 = mean_squared_error(y_valid, y_pred_valid2)

plt.bar(np.array([error_train1, error_valid1]), np.array([error_train2, error_valid2]))
plt.show()



#Lasso 0
model1 = Lasso(alpha=0).fit(X_train, y_train)
y_pred_train1 = model1.predict(X_train)
y_pred_valid1 = model1.predict(X_valid)
error_train1 = mean_squared_error(y_train, y_pred_train1)
error_valid1 = mean_squared_error(y_valid, y_pred_valid1)

#Lasso 0.8
model2 = Lasso(alpha=0.8).fit(X_train, y_train)
y_pred_train2 = model2.predict(X_train)
y_pred_valid2 = model2.predict(X_valid)
error_train2 = mean_squared_error(y_train, y_pred_train2)
error_valid2 = mean_squared_error(y_valid, y_pred_valid2)

plt.bar(np.array([error_train1, error_valid1]), np.array([error_train2, error_valid2]))
plt.show()
```
